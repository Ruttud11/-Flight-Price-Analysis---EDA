```python
#deal with outliers

def replace_outliers(df, col):
    Q1= df[col].quantile(0.25)
    Q2= df[col].quantile(0.50)
    Q3= df[col].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+ 1.5*IQR
    df[col] = np.where((df[col] < lower_bound) | (df[col]>upper_bound),Q2,df[col])
    return df

cols_with_outliers = ['duration','price']
for col in cols_with_outliers:
    df = replace_outliers(df, col)
    
    
```

```python
n_cols = ["duration",'days_left',"price"]
plt.figure(figsize=(12,6))

for i, col in enumerate(n_cols,1):
    plt.subplot(2,2,i)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.show()    
```

```python
route_counts = df.value_counts(["source_city","destination_city"]).reset_index(name="Flight_count")
route_counts["route"] = route_counts["source_city"] + " -> "+route_counts['destination_city']
route_counts
                                                                    
```

```python
route_counts.nlargest(5,"Flight_count")
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(data = route_counts,x= "Flight_count",y="route",hue="route",palette= "Set1")
for bar in ax.containers:
    ax.bar_label(bar)

plt.title("Flight Routes Distiribution")
plt.show()
```

```python
sourceDepar = df.value_counts(["source_city","departure_time"]).reset_index(name="Flight_count")
sourceDepar["sourceDepar"] = sourceDepar["source_city"] + " -> "+sourceDepar['departure_time']
sourceDepar
                                                                    
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(data = sourceDepar,x= "Flight_count",y="sourceDepar",hue="sourceDepar",palette= "Set1")
for bar in ax.containers:
    ax.bar_label(bar)

plt.title("Flight Count by Source City and Departure Time")
plt.show()
```
